```python
# visuallization of change of condition acc to temperature 
plt.figure(figsize=(15,6))
sns.barplot(x='Weather Condition', y='Temp_C', data=d)
plt.xticks(rotation=90)
plt.show
```

```python
# visualization of wind speed acc. to each weather condition
plt.figure(figsize=(15,6))
sns.barplot(y='Wind Speed_km/h',x='Weather Condition',data=d)
plt.xticks(rotation=90)
plt.show
```

```python
# boxplot to see in which range most of temperature of weather falls into
plt.figure(figsize=(15,10))
plt.boxplot(x ='Temp_C', data=d)
```
